# Enhancing the AI Quest with Database Features

## Project Overview

For our improvements of the quest from Trimester 1 of Computer Science, we've decided to build off the **AI Submodule** from the Digital Famine Quest. We're adding six database-driven features to make the quest more interactive, collaborative, and data-informed.

These features will use **SQLite databases** to store player information, track progress, and create a shared learning experience across the entire class.

## The Six Features

1. **Leaderboard** - Display top player scores
2. **Feedback Survey** - Collect and show player feedback
3. **Quiz Analytics** - Track which questions students struggle with most
4. **Vault Code Storage** - Auto-save mission completion codes
5. **Progress Tracking** - Visual progress bars showing completion status
6. **Recent Prompts Display** - Show the latest 3 prompts from other players

---

## Feature 1: Leaderboard 

### Visual Design
A table at the top displaying the 10 highest scores 
- Rank position (1-10)
- Player username 
- Total score (quizzes + games) 

Makes the game more engaging and interactive as people try to beat each others high scores.

### Database Integration
The database stores each player's name and score who played the game or took the quiz. When someone completes all of the topics and the respective quiz or game for that section, their score updates automatically. The leaderboard queries the database to find the top 10 highest scores and displays them in descending order. 

---

## Feature 2: Feedback Survey 

### Visual Design
A two-part interface:
- **Top section**: Feedback form with star ratings (1-5), category dropdown (Bug/Suggestion/Praise), and text box
- **Bottom section**: Recent feedback cards showing previous responses.

### Database Integration
When students finish the quest, they can submit feedback which gets saved to the database. The system stores their rating, category, comments, and timestamp. Below the form, it displays the 10 most recent feedback entries from other students, creating transparency and showing that feedback is valued.

---

## Feature 3: Quiz Analytics 

### Visual Design
A data dashboard showing:
- Bar chart ranking questions by difficulty (based on class success rates)
- Color-coded difficulty indicators: 🟢 Easy (>80% correct), 🟡 Medium (60-80%), 🔴 Hard (<60%)
- Each question shows: question text, success rate percentage, and number of attempts
- Class average score displayed prominently at the top

### Database Integration
Every time a student answers a quiz question, the database records whether they got it right or wrong. The system calculates what percentage of the class answered each question correctly. This shows which concepts the entire class struggles with most, helping teachers identify topics that need more explanation. Students see class-wide data, not individual performance, reducing pressure.

---

## Feature 4: Vault Code Storage 

### Visual Design
A secure vault interface with three code slots:
- Each slot represents a mission (Mission 1, 2, 3)
- Locked slots appear dark and grayed out with a 🔒 icon
- Unlocked display the code in large text
- A final section combines all three codes into the complete vault PIN

### Database Integration
When students complete a mission, their unique code is automatically saved to the database instead of requiring them to write it down on paper. Students can view their codes anytime by logging in. The system checks if all three codes are collected before allowing vault access, preventing lost or forgotten codes and reducing frustration.

## Feature 5: Progress Tracking

### Visual Design
A progress bar at the main page with all of our submodules: digital-famine/ai/

- Large animated progress bar at the top showing overall completion (0-100%) of the modules
- Change the background of the submodule and make the progress bar fit into the color theme of the background
- Estimated time to complete the AI module near progress bar
- After every submodule is completed (4 total) a green check box will appear near the title of the submodule

### Database Integration
The database tracks which sections have already been completed. Every time a section is completed, it marks that section as done as well as fills in the progress bar which shows how close they are to completing the module. The system calculates the percentage complete by dividing completed sections by total sections (4 sections).Even when  players close their browser and return later, the database remembers exactly where they left off, so they do not lose progress.

## Feature 6: Recent Prompts Display 

### Visual Design
A community prompt feed in the AI interaction section:
- Shows three recent prompts
- Each card displays: player name, the prompt text
- "Try this prompt" button to copy text
- Auto-refreshes to show new prompts

Located right below where students type their own AI prompts.

### Database Integration
When students submit a prompt to the AI, it gets saved in the database along with a timestamp. The display area queries for the three most recent prompts from all students and shows them. This helps students who are stuck see examples of what questions others are asking, fostering collaboration and giving ideas for how to interact with the AI effectively.

# Enhancing the AI Quest with Database Features

## Project Overview

For our improvements of the quest from Trimester 1 of Computer Science, we've decided to build off the **AI Submodule** from the Digital Famine Quest. We're adding six database-driven features to make the quest more interactive, collaborative, and data-informed.

These features will use **SQLite databases** to hold information of the user, track progress, and create a shared learning experience across the entire class.

## The 6 Features

1. **Leaderboard** - Display top player scores
2. **Feedback Survey** - Collect and show player feedback at the end
3. **Quiz Analytics** - Track which questions students struggle with most
4. **Vault Code Storage** - Auto-save mission completion codes
5. **Recent Prompts Display** - Show the latest 3 prompts from other players
6. **Jokes** - Display funny jokes in between questions

## Feature 1: Leaderboard 


### Visual Design
A table at the top displaying the 10 highest scores 
- Rank position (1-10)
- Player username 
- Total final score (including quizzes and games) 

Makes the game more engaging and interactive as people try to beat each others high scores. Adds a competitive element and comparative element to see how you are doing compared to other players!

### Database Integration
The database stores each player's name and score who played the game or took the quiz. When someone completes all of the topics and the respective quiz or game for that section, their score updates automatically. The leaderboard queries the database to find the top 10 highest scores and displays them in descending order. The other scores are cut off. This involves using the continuous updating aspect of databases as new information is continuously added to the database and we extract updated information from the database to update the leaderboard.

## Feature 2: Feedback Survey 

### Visual Design
A two-part interface:
- **Top section**: Feedback form with star ratings (1-5), category dropdown (Bug/Suggestion/Praise), and text box
- **Bottom section**: Recent feedback cards showing previous responses.

### Database Integration
At the end of the quest students will be given a survey they can complete and they can submit feedback which gets saved to the database. The system stores their rating, category, comments, and timestamp. Below the form, it displays the 10 most recent feedback entries from other students, creating transparency and showing that feedback is valued.

## Feature 3: Quiz Analytics 

### Visual Design
A data dashboard showing:
- Bar chart ranking questions by difficulty (based on class success rates)
- Color-coded difficulty indicators: 🟢 Easy (>80% correct), 🟡 Medium (60-80%), 🔴 Hard (<60%)
- Each question shows: question text, success rate percentage, and number of attempts
- Class average score displayed prominently at the top

### Database Integration
Every time a student answers a quiz question, the database records whether they got it right or wrong. The system calculates what percentage of the class answered each question correctly. This shows which concepts the entire class struggles with most, helping teachers identify topics that need more explanation. Students see class-wide data, not individual performance, reducing pressure.

## Feature 4: Vault Code Storage 

1.  Purpose
   The autofill system automatically saves and displays mission completion codes so students never need to write them down or enter them manually. It prevents lost codes and makes the vault feel like a secure, polished feature.

2.  Database Details
   **Table:** `vault_codes`  
   This table stores one row per user per mission:

   | Field | Meaning |
   |-------|---------|
   | `user_id` | The student using the quest |
   | `mission_number` | Mission 1, 2, or 3 |
   | `code` | The unique mission code |

   Whenever a mission is completed, the system immediately stores or updates the code in the database.

3.  How Autofill Works in the Vault
   When the Vault screen loads, the system checks the database for the student's saved mission codes.

   - **If a code exists:** The slot unlocks and shows the code.  
   - **If no code exists:** The slot stays gray with a 🔒 “Locked” label.  
   - **If all three codes exist:** The final vault PIN appears (code1–code2–code3).

   No manual input is required from the student at any point.

4.  Student Experience
   1. Complete a mission → code is saved automatically.  
   2. Open the vault → unlocked slots appear instantly.  
   3. Complete all missions → the final PIN is revealed.

## Feature 5: Recent Prompts Display 

### Visual Design
A community prompt feed in the AI interaction section:
- Shows three recent prompts
- Each card displays: player name, the prompt text
- "Try this prompt" button to copy text
- Auto-refreshes to show new prompts

Located right below where students type their own AI prompts.

### Database Integration
When students submit a prompt to the AI, it gets saved in the database along with a timestamp. The display area queries for the three most recent prompts from all students and shows them. This helps students who are stuck see examples of what questions others are asking, fostering collaboration and giving ideas for how to interact with the AI effectively.

## Feature 6: Joke

When the user completes 3 questions, to give them some motivation to keep progressing forward through the quest and add a fun element, we will display a funny computer science joke on the screen by using databases.

Eg. "Why Do Programmers Prefer Dark Mode?

Because light attracts bugs!"

### Database Integration

A fixed set of jokes will be stored in the database and users can add more jokes to that database at the end with the survey. We will then randomly display jokes from the new database after the user answers 3 questions.
